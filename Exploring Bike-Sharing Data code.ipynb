#### Task2.	Visualize the temperature by day by using appropriate graph. 

```python
import matplotlib.pyplot as plt

day['dteday'] = pd.to_datetime(day['dteday'])

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(day['dteday'], day['temp'], marker='o', linestyle='-')
plt.title('Temperature by Day')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(day['temp'], bins=20, color='skyblue', edgecolor='black')
plt.title('Temperature Distribution')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### Task#3:Show the shape of the data by using histogram for the number of rentals per day in 2012. 

```python
import matplotlib.pyplot as plt

# Filter data for the year 2012
rentals_2012 = day[day['yr'] == 1]
#rentals_2012 = data[data['mnth'] == 2012]
# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(rentals_2012['cnt'], bins=20, color='orange', edgecolor='black')
plt.title('Distribution of Rentals per Day in 2012')
plt.xlabel('Number of Rentals')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### TASK# 4 :Make boxplot of the total number of rentals by months in 2012.  

```python

# Filter data for the year 2012
rentals_2012 = day[day['yr'] == 1]
#looping the data to get each month if we dont do that we get sum of all data:
monthly_rentals = []
for month in range(1, 13):
    rentals_month = rentals_2012[rentals_2012['mnth'] == month]['cnt']
    monthly_rentals.append(rentals_month)
boxprops = dict(facecolor='lightblue', color='blue')
whiskerprops = dict(color='blue', linestyle='-')
medianprops = dict(color='red', linewidth=2)
flierprops = dict(marker='o', markerfacecolor='gray', markersize=8, linestyle='none')    
# Create a boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(monthly_rentals, patch_artist=True, boxprops=boxprops, whiskerprops=whiskerprops, medianprops=medianprops, flierprops=flierprops)
plt.title('Total Number of Rentals by Month in 2012',fontsize=16)
plt.xlabel('Month',fontsize=14)
plt.ylabel('Total Rentals',fontsize=14)
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### Task#5:Visualize the relationship between temperature and bicycle rentals and write in your own words. 

```python
import matplotlib.pyplot as plt

# Plotting temperature vs. bicycle rentals
plt.figure(figsize=(13, 6))
plt.scatter(day['temp'], day['cnt'], color='darkblue', alpha=0.7)
plt.title('Relationship between Temperature and Bicycle Rentals')
plt.xlabel('Temperature (C)')
plt.ylabel('Bicycle Rentals')

plt.text(0.5, 0.97 ,"The plot suggests a positive correlation between temperature and rental demand.\nHigher temperatures tend to coincide with increased bicycle rentals.",
         horizontalalignment='right', verticalalignment='top', transform=plt.gca().transAxes, fontsize=10, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

plt.text(0.5, 0.03, "However, variability in rental numbers at similar temperature levels\nindicates that other factors may also influence rental behavior.",
         horizontalalignment='left', verticalalignment='bottom', transform=plt.gca().transAxes, fontsize=10, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0'))
plt.grid(True)
plt.tight_layout()
plt.show()
```
